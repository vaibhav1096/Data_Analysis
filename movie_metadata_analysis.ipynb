## GENRE BASED ANALYSIS

```python
#working on Genres collumn
df_['genres']
```

```python
pd.Categorical(df_['genres'])
```

```python
l=[]
for i in df_['genres']:
    a=i.split('|')
    for j in a:
        if j in l:
            pass
        else:
            l.append(j)
print(l)
```

```python

GenreDF = pd.DataFrame(columns=['genre','imdb_score','country','budget','gross'])
```

```python
GenreDF
```

```python
#Creatig new dataframe based on categories
i = 0
values = []
for idx, row in df_.iterrows():
    genres_val = row['genres']
    if not pd.isnull(genres_val) and not pd.isnull(row['gross']):
        for genre in genres_val.split('|'):
            values.append([ genre,row['imdb_score'],row['country'],row['budget'],row['gross']])  
            
GenreDF = pd.DataFrame(values, columns=['genre','imdb_score','country','budget','gross'])
```

```python
GenreDF['profit']=GenreDF['gross']-GenreDF['budget']
```

```python
GenreDF.describe()
```

```python
#Average IMDB score for diffrent genres
avg_score=GenreDF.groupby(by='genre')[['genre','imdb_score']].mean()
avg_score.reset_index().sort_values(['imdb_score'],ascending=False).head()
```

```python
#Highest budgeted genre
pd.set_option('display.float_format', lambda x: '%.3f' % x)
High_budget=GenreDF.groupby(by='genre')[['genre','budget']].mean()
```

```python
High_budget.reset_index().sort_values(['budget'],ascending=False).head()
```

```python
#Highest profitable genre
avg_score=GenreDF.groupby(by='genre')[['genre','profit']].mean()
avg_score.reset_index().sort_values(['profit'],ascending=False).head()
```

```python
GenreDF['year']=df_['title_year']
```

```python
yw=GenreDF.groupby(['genre','year'])[['profit']].mean()
```

```python
for genre, new_df in yw.groupby(level=0):
    print(genre)
    plt.figure()
    sns.lineplot(x=new_df.reset_index().drop(columns='genre')['year'],y=new_df.reset_index().drop(columns='genre')['profit'])
    plt.show()
```

## COUNTRY BASED ANALYSIS

```python
countryDf=GenreDF
```

```python
countryDfgb=countryDf.groupby(['country','genre']).count()
countryDfgb
```

```python
for country, new_df in countryDfgb.groupby(level=0):
    a=new_df.reset_index()['imdb_score']
    label_lis=new_df.reset_index()['genre'].values
    print(country)
    plt.figure()
    plt.pie(a,labels=label_lis)
    plt.show()
```

```python
countryDf1=df_.groupby(['country'])
```

```python
a=countryDf1[['budget','gross']].count().reset_index()
```

```python
f = plt.figure()
f.set_figwidth(9)
f.set_figheight(1)
sns.barplot(x=a['country'],y=a['budget'])
```

```python
a.sort_values('budget',ascending=False).head()
```

```python
b=countryDf1[['budget','gross']].mean().reset_index()
```

```python
b['profit %']=((b['budget']-b['gross'])/b['budget'])*100
```

```python
b.sort_values(['profit %'],ascending=False)[['country','profit %']].head()
```
